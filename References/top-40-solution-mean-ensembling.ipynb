# Library Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import statistics
```

# Reading in Submissions

```python
sub = pd.read_csv('/kaggle/input/playground-series-s3e17/sample_submission.csv')
sub1 = pd.read_csv('/kaggle/input/machine-failure-ensemble-dataset/Submission/sub1.csv')
sub2 = pd.read_csv('/kaggle/input/machine-failure-ensemble-dataset/Submission/sub2.csv')
sub3 = pd.read_csv('/kaggle/input/machine-failure-ensemble-dataset/Submission/sub2.csv')
```

```python
# Mean ensembling weights
weights = [0.8, 0.15, 0.05]
```

# Submission EDA

```python
plt.figure(figsize=(4,3),dpi=96)
plt.title('Sub1 Prediction Visualization')
plt.xlabel('Machine Failure Prediction')
plt.ylabel('# of Occurences')
plt.hist(sub1['Machine failure'], bins=16, log=True, color='b')
plt.show()
```

```python
plt.figure(figsize=(4,3),dpi=96)
plt.title('Sub2 Prediction Visualization')
plt.xlabel('Machine Failure Prediction')
plt.ylabel('# of Occurences')
plt.hist(sub2['Machine failure'], bins=16, log=True, color='b')
plt.show()
```

```python
plt.figure(figsize=(4,3),dpi=96)
plt.title('Sub3 Prediction Visualization')
plt.xlabel('Machine Failure Prediction')
plt.ylabel('# of Occurences')
plt.hist(sub3['Machine failure'], bins=16, log=True, color='b')
plt.show()
```

# Mean Ensembling

```python
sub['Machine failure'] = (weights[0]*sub1['Machine failure'] + weights[1]*sub2['Machine failure'] + weights[2]*sub3['Machine failure'])/3
```

```python
sub.to_csv('submission.csv', index=False)
```

# Submission Visualization

```python
plt.figure(figsize=(4,3),dpi=96)
plt.title('Final Submission Prediction Visualization')
plt.xlabel('Machine Failure Prediction')
plt.ylabel('# of Occurences')
plt.hist(sub['Machine failure'], bins=16, log=True, color='b')
plt.show()
```
